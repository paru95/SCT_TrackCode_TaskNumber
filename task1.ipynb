## Extarction of top 10 countries with highest male and female population

```python
males_population_data = df[df['Series Code'] == 'SP.POP.TOTL.MA.IN']
males_population_sorted = males_population_data.sort_values(by='2022', ascending=False)
males_top_ten_countries = males_population_sorted.head(10)

print("Top Ten Countries of Male Population")
print(males_top_ten_countries[[ 'Country Code']])
```

```python
female_population_data = df[df['Series Code'] == 'SP.POP.TOTL.FE.IN']
female_population_sorted = female_population_data.sort_values(by='2022', ascending=False)
female_top_ten_countries = female_population_sorted.head(10)

print("Top Ten Countries of female Population")
print(female_top_ten_countries[[ 'Country Code']])
```

## Top ten countries with highest male and female population in 2022

```python
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code",data = males_top_ten_countries , palette = "viridis")
plt.title("top ten countries of males population data (2022)", fontsize=10)
plt.xlabel("male population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2022",y="Country Code",data = female_top_ten_countries, palette = "viridis")
plt.title(" top ten countries of males population data (2022))",fontsize=10)
plt.xlabel("female population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

## Stacked Bar Plot

## Top 10 countries with male and female population

```python
merge_data = pd.merge( male_population_data,female_population_data,on="Country Code",suffixes=("_male", "_female"))
```

```python
merge_data["Total population"] = merge_data["2022_male"] + merge_data ["2022_female"]
```

```python
merge_data. head ()
```

```python
sorted_data = merge_data.sort_values(by="Total population", ascending=False)
```

## Select the top 10 countries with the highest total population

```python
top_10_countries = sorted_data.head(10)
country_codes = top_10_countries["Country Code"]
female_pop = top_10_countries["2022_female"]
male_pop = top_10_countries["2022_male"]


plt.figure(figsize=(12,6))
plt.bar(country_codes, female_pop, color="red", label="Female Population")
plt.bar(country_codes, male_pop, bottom=female_pop, color="green", label="Male Population")

plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries by Total Population (Stacked Male & Female)")
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```

## Bottom 10 countries with male and female population(2022)

```python
bottom_10_countries = sorted_data. tail (10)
```

```python
country_codes = bottom_10_countries["Country Code"]
female_pop = bottom_10_countries["2022_female"]
male_pop = bottom_10_countries["2022_male"]


plt.figure(figsize=(12,6))
plt.bar(country_codes, female_pop, color="red", label="Female Population")
plt.bar(country_codes, male_pop, bottom=female_pop, color="green", label="Male Population")

plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries by Total Population (Stacked Male & Female)")
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```
